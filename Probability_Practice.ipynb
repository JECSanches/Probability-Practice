# Probabilidade (Prática)

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import stats
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Definindo as variáveis
mu = 15
sigma = .1

# Os valores limites para análise
low = 15.2
high = np.inf

# Definindo os eixos
x = np.linspace(mu - 4*sigma, mu + 4*sigma, num=500)
y = stats.norm.pdf(x, mu, sigma) # Função densidade de probabilidade

# Definindo a região de probabilidade
prob_area = stats.norm.cdf(high, mu, sigma) - stats.norm.cdf(low, mu, sigma)
print(f'--> A probabilidade de um pacote com mais de 15.2 Kg ser selecionado é de {prob_area:.2%}\n')

fig, ax = plt.subplots(1, 1, figsize=[10,6])
ax.set_title('Demarcação da região de probabilidade')
ax.plot(x, y, color='blue')
ax.fill_between(x, y, where= (low < x) & (x < high), alpha=.5)
ax.grid()
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Para tal façamos uso do ppf (percent point function)

# Valores de média, desvio padrão e percentil
mu = 15
sigma = .1
percentil = .95 # lower tail probability

value = stats.norm.ppf(percentil, mu, sigma)
print(f'--> Cerca de 95% dos pacotes produzidos têm massa menor que {value:.2f}kg')

# Os valores limites para análise
low = - np.inf
high = value

# Definindo os eixos
x = np.linspace(mu - 4*sigma, mu + 4*sigma, num=500)
y = stats.norm.pdf(x, mu, sigma) # Função densidade de probabilidade

fig, ax = plt.subplots(1, 1, figsize=[10,6])
ax.set_title(f'Demarcação da região de probabilidade correspondente a {percentil:.2%}')
ax.plot(x, y, color='blue')
ax.fill_between(x, y, where= (low < x) & (x < high), alpha=.5)
ax.grid()
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y = np.linspace(-6, 6, 200)
```

```python
# Distribuição normal Y1

# Os parâmetros
mu = 0 # Média
variance = 1 # Variância
sigma = math.sqrt(variance) # Desvio padrão

normal = stats.norm.cdf(Y, mu, sigma)
normal
```

```python
# Distribuição t-student - Y2
gl = 5 # graus de liberdade

t_student = stats.t.cdf(Y, gl)
t_student
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
plt.figure(figsize=[10,5])
plt.plot(Y, normal, label='Y1 - Normal distribution')
plt.plot(Y, t_student, label='Y2 - t-student distribution')
plt.legend()

plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
```

```python
# Definindo
mu = 0 # Média
variance = 1 # Variância
sigma = math.sqrt(variance) # Desvio padrão
gl = 5 # Graus de liberdade (para Y2)

# Calculando a distribuição normal - Y1
Y1 = stats.norm.ppf(Y, mu, sigma)
# Calculando a distribuição t-student - Y2
Y2 = stats.t.ppf(Y, gl)
```

```python
print('Distribuição normal Y1: \n', Y1, '\n')
print('Distribuição t-student Y2: \n', Y2)
```

6. Faça um gráfico com os dados do item anterior.

```python
plt.figure(figsize=[10,5])
plt.plot(Y, Y1, label='Y1 - Distribuição normal')
plt.plot(Y, Y2, label='Y1 - Distribuição t-student')
plt.legend()

plt.show()
```
